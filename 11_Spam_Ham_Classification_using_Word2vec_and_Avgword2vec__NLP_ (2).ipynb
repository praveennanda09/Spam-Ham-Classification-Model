```python
import pandas as pd

# Path to the downloaded file
file_path = '/content/drive/MyDrive/Data Science - GT/Natural Language Processing/SMSSpamCollection.txt'

# Load the file with options to handle inconsistent data
df = pd.read_csv(
    file_path,
    sep='\t',
    names= ["label", "messages"])

# Show the first few rows of the DataFrame
print(df.head())
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['messages'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, df['messages']) if i<1]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
# Function to calculate average Word2Vec vector for a document
def avg_word2vec(doc):
    # Get the dimensionality of Word2Vec vectors
    vector_size = model.wv.vector_size

    # Calculate the mean of word vectors, or return a zero vector if no valid words
    if len([word for word in doc if word in model.wv.index_to_key]) > 0:
        return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
    else:
        return np.zeros(vector_size)  # Return zero vector if no valid words
```
